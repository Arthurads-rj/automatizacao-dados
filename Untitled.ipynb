# Automação de Processos com Análise de dados

Nesse projeto, será demonstrado um exemplo de aplicação desenvolvida com o intuito de automatização de análise de dados.

Esse projeto fará 6 passos:
1. Irá entrar no sistema (exemplificado com o Google Drive)
2. Acessar o local do relatório
3. Baixar o relatório
4. Analisar os dados e calcular o que foi programado
5. acessar o email
6. enviar o resultado para o email programado

Todas essas coisas seriam feitas de maneira manual, mas é possivel automatizar essas tarefas.
As tarefas utilizaram de arquivos que eu criei para exemplificar o projeto, além de acessar meu Google Drive onde está hospedado o arquivo.

Primeiro, é preciso instalar o **pyautogui** e o **pyperclip**. O pyperclip já vem com a instalação do pyautogui.

```python
#pip install pyautogui
```

```python
import pyautogui

import pyperclip

from time import sleep #necessário para conseguir estabelecer pausas entre algumas atividades
pyautogui.PAUSE = 2.5


# 1. entrar no sistema
pyautogui.hotkey('ctrl', 't') #criar uma guia nova
pyperclip.copy('https://drive.google.com/drive/u/1/folders/1J_s6oHdsW9N2u675Snltagzr7ChXz9Z-') #copiar o link do google drive
pyautogui.hotkey('ctrl', 'v') #colar o link
#2.Acessar o local do relatório
pyautogui.press('enter') #acessar o google drive
sleep(5)

#3. Baixar o relatório
pyautogui.click(x=387, y=343)
pyautogui.click(x=1153, y=150)
pyautogui.click(x=1019, y=565)
sleep(5)

```

A partir disso, o programa irá executar algumas linhas de código que irão calcular alguns dados do próprio relatório.
normalmente, o arquivo é baixado para a pasta downloads, mas o arquivo ficará na pasta principal do programa para facilitar o funcionamento e entendimento de como ele funciona.

Esses dados são:
- O faturamento
- Quantidade de produtos vendidos

Após isso, também será enviado por email os dados coletados

```python
#4. Analisar os dados
import pandas as pd

tabela = pd.read_excel('Vendas - Mar.xlsx') #caso baixado deveria estar como r'C:\Users\usuario\Downloads\Vendas - Mar.xlsx'

display(tabela)

faturamento = tabela['Valor Final'].sum()
quantidade = tabela['Unidade'].sum()

#5. Acessar o email
pyautogui.hotkey('ctrl', 't')
pyperclip.copy('https://mail.google.com/mail/u/1/#inbox')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
sleep(7)

#6. Enviar o resultado para o email programado

pyautogui.click(x=116, y=174)
sleep(2)
pyautogui.write('artlops.237+diretoria@gmail.com')
pyautogui.press('tab') #seleciona o email
pyautogui.press('tab') #vai pro próximo campo
pyperclip.copy('Relatório de vendas')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

texto = f'''
Bom dia,

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}
'''

pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')
#pyautogui.hotkey('ctrl', 'enter')
```

Só para exemplificar como fica no final de todo o procedimento



depois disso ele envia para o email o relatório e assim se conclui o uso do programa.
